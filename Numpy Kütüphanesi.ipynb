```python
#korelasyon hesabı
np.random.seed(42)#random state 42 kullanılır 
#böylece sonuçların aynı şekilde tekrarlanması sağlanır
notlar1=np.random.normal(loc=50,scale=5,size=20)
notlar2=.70*notlar1+np.random.normal(loc=0,scale=5,size=20)
corr=np.corrcoef(notlar1,notlar2)
corr
```

```python
#Cihan korelasyon hesabı
np.random.seed(42) #random state 42 kullanılır bu şekilde sonuçların aynı şekilde tekrarlanması sağlanır
size=20
notlar1=np.random.normal(loc=50, scale=5,size=20)
notlar=.50*notlar1+np.random.normal(loc=10,scale=5,size=20)
corr=np.corrcoef(notlar1,notlar2)
corr
```
